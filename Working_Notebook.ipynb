```python
import pandas as pd
from textblob import TextBlob
import re
from nltk.stem import WordNetLemmatizer
import csv
```

```python
df = pd.read_csv('data/TyrionLannister_Tweets.csv')
df.head()
```

```python
updated_df = df['text']
updated_df.head()
```

```python
tweet_text = []
tweet_timestamp = []
tweet_date = []
tweet_char = []
tweet_sen = []
tweet_pol = []
tweet_sub = []



for key, tweet in updated_df.items():
        analysis = TextBlob(tweet)

        senti = ""
        polarity = 0.0
        subjectivity = 0.0
    
        if analysis.sentiment[0]>0:
           senti = "Pos"
        elif analysis.sentiment[0]<0:
           senti = "Neg"
        else:
           senti = "Neu"
        polarity = float(analysis.sentiment[0])
        subjectivity = float(analysis.sentiment[1])

        tweet_text.append(tweet)
        tweet_pol.append(polarity)
        tweet_sen.append(senti)
        tweet_sub.append(subjectivity)
#         print(tweet_text)
```

```python
df = pd.DataFrame({
  "text": tweet_text,
  "sentiment" : tweet_sen,
  "polarity" : tweet_pol,
  "subjectivity" : tweet_sub,
})

df.head()
```

```python
df.to_csv("data/TyrionLannister_Sentiments.csv", index=False)
```

```python
df = df.dropna()
stemmer = WordNetLemmatizer()

for index, row in df.iterrows():
    # Remove all the special characters
    test_preproc = re.sub(r'\W', ' ', str(row["text"]))

    # remove all single characters
    test_preproc = re.sub(r'\s+[a-zA-Z]\s+', ' ', test_preproc)

    # Remove single characters from the start
    test_preproc = re.sub(r'\^[a-zA-Z]\s+', ' ', test_preproc) 

    # Substituting multiple spaces with single space
    test_preproc = re.sub(r'\s+', ' ', test_preproc, flags=re.I)

    # Removing prefixed 'b'
    test_preproc = re.sub(r'^b\s+', '', test_preproc)

    # Converting to Lowercase
    test_preproc = test_preproc.lower()
    
    # Lemmatization
    test_preproc = test_preproc.split()

    test_preproc = [stemmer.lemmatize(word) for word in test_preproc]
    comb_string = ' '.join(test_preproc)
    df.loc[index, "Preprocess"] = comb_string

```

```python
df.to_csv("data/SansaStark_Sentiments.csv", index=False)
```
